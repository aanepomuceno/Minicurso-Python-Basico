```python
import numpy as np
import itertools
import matplotlib.pyplot as plt
```

Questão 4.1

```python
#importar dados de dengue (DATA SUS)
f_dengue = open('casos_dengue_rio_2019.csv',encoding="ISO-8859-1")
acumulado = 0
print("Ano: 2019")
print(f'{"Mês":14}',f'{"Número de Casos":16}',"Acumulado")
print('-'*41)
for line in itertools.islice(f_dengue,7,19):
    lista_data = line.split(';')
    acumulado += float(lista_data[1])
    print("{:14}".format(lista_data[0][1:-1]),"{:16}".format(lista_data[1][:-1]),int(acumulado))
f_dengue.close()
#o slice [1:-1] é usado para retirar as aspas dos nomes dos meses. O slice [:-1] retira o '\n' dos dados 
#da segunda coluna, evitando que ele pule uma linha.
```

Questão 4.2

```python
C = 1 #m/s2
beta = [0.1,0.2,0.3]  #s-1
omega_0 = 2.0
omega = np.linspace(0.5,4,100)
```

```python
#vamos plotar as funções
for k in beta:
    amplitude = C/( np.sqrt( (omega_0**2-omega**2)**2+4*omega**2*k**2 ))
    plt.plot(omega,amplitude,label=r"$\beta$="+str(k))
plt.xlabel(r"$\omega [s^{-1}]$",fontsize=12)
plt.ylabel("Amplitude [m]",fontsize=12)
plt.legend()
```

Questão 4.3

```python
#importar os dados
viral_data = np.loadtxt("HIVseries.csv",delimiter=",")
time_data = viral_data[:,0]
carga_data = viral_data[:,1]
print(time_data.min())
print(time_data.max())
```

```python
#modelo
time = np.linspace(0,10,100)
A,B,alpha,beta = 80000,80000,0.5,0.35
viral_load = A*np.exp(-alpha*time) + B*np.exp(-beta*time)
```

```python
#plot dados e modelo
plt.plot(time_data,carga_data,"k.",markersize=8, label="Dados")
plt.plot(time,viral_load,'r-',label="Modelo")
plt.xlabel("Tempo [dias]",fontsize=12)
plt.ylabel("Carga Viral",fontsize=12)
plt.legend(frameon=False,fontsize=12)
```
